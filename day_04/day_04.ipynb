```python
# f = open("./test.txt", "r")
f = open("./input.txt", "r")

running_sum = 0
# lines = []

for l in f:
    l = l.strip()
    card, nums = l.split(":")
    card_num = card.split(" ")[1]
    winning_nums, have_nums = nums.split("|")
    winning_nums = [ int(i) for i in winning_nums.split(" ") if i!='']
    have_nums = [ int(i) for i in have_nums.split(" ") if i!='']

    matches = -1
    val = 0

    for x in have_nums:
        if x in winning_nums:
            matches += 1
    
    if matches >= 0:
        val = 2 ** matches
    
    running_sum += val

    print(card_num, "-", winning_nums, "-", have_nums, "=", val, running_sum)

print("Running sum:",running_sum)
```

```python
# f = open("./test.txt", "r")
f = open("./input.txt", "r")

running_sum = 0
cards = []


for l in f:
    l = l.strip()
    card, nums = l.split(":")
    card_num = [ int(i) for i in card.split(" ")[1:] if i!=''][0]
    winning_nums, have_nums = nums.split("|")
    winning_nums = [ int(i) for i in winning_nums.split(" ") if i!='']
    have_nums = [ int(i) for i in have_nums.split(" ") if i!='']
    cards.append({"id": card_num, "win":winning_nums, "have": have_nums, "count": 1})

# print(cards)

for card in cards:
    matches = 0
    val = 0

    for x in card["have"]:
        if x in card["win"]:
            matches += 1
    
    print(card["id"], card["count"], "\t", card["win"], "\t", card["have"], "\t", matches, " \t ")

    if matches >= 0:
        for i in range(matches):    
            print(card["id"]+i+1, "-> +",card["count"])
            next(item for item in cards if item["id"] == card["id"]+i+1)["count"] += card["count"]

for card in cards:
    running_sum += card["count"]

print("Running sum:",running_sum)
```
