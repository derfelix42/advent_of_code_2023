```python
# f = open("./test.txt", "r")
f = open("./input.txt", "r")

running_sum = 0

# A lot of parsing the line to get to the good stuff...
for l in f:
    l = l.strip()
    game = l.split(":")
    game_num = game[0].split(" ")[1]
    game_val = int(game_num)
    plays = game[1].split(";")
    print("Game",game_num)
    for play in plays:
        for cube in play.split(","):
            cube = cube.strip()
            num, color = cube.split(" ")
            print(" -",color, num, end="")

            # Here is the good stuff, just check if any number of cubes are too many for each color
            if color == "red" and int(num) > 12:
                game_val = 0
            if color == "green" and int(num) > 13:
                game_val = 0
            if color == "blue" and int(num) > 14:
                game_val = 0
        print()
    
    # The gameval got set to the ID of the game initially, but might have been set to 0, if the game is impossible.
    if game_val == 0:
        print("-> This game is impossible!")
    running_sum += game_val

    print()

print("Running sum:",running_sum)
```

```python
# f = open("./test.txt", "r")
f = open("./input.txt", "r")

running_sum = 0

# Again same parsing to get to the good stuff
for l in f:
    l = l.strip()
    game_num = l.split(":")[0].split(" ")[1]
    red_max = 0
    green_max = 0
    blue_max = 0
    plays = l.split(":")[1].split(";")
    for play in plays:
        # print(play)
        for cube in play.split(","):
            cube = cube.strip()
            num, color = cube.split(" ")
            # print(color, num)

            # simply find the maximum for each colored cube in each game
            if color == "red" and int(num) > red_max:
                red_max = int(num)
            if color == "green" and int(num) > green_max:
                green_max = int(num)
            if color == "blue" and int(num) > blue_max:
                blue_max = int(num)

    # calculate the value of the game and add to the running sum
    game_val = red_max * blue_max * green_max
    print(game_num,"- gamevalue:",red_max,"*",green_max,"*",blue_max,"=",game_val)
    
    running_sum += game_val

print("Running sum:",running_sum)
```
